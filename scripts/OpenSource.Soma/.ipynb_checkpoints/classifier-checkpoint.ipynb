# data cleaning

```python
def check_missing_values(X, y): 
    print("Checking for missing values in Features (X) and Target (y):\n")
    print("Missing values in Features (X):")
    print(X.isnull().sum())

    print("\nMissing values in Target (y):")
    print(y.isnull().sum())

    print("\nIdentifying potential outliers:")
    Q1 = X.quantile(0.25)
    Q3 = X.quantile(0.75)
    IQR = Q3 - Q1
    
    # Flagging outliers
    outliers = ((X < (Q1 - 1.5 * IQR)) | (X > (Q3 + 1.5 * IQR))).sum()
    print("Number of potential outliers per feature:")
    print(outliers)
```

```python
print("Train data check: ")
check_missing_values(X_train, y_train)
```

### Training an LSTM model for time series prediction

```python
train_data["timestamp"] = pd.to_datetime(train_data["timestamp"])
train_data = train_data.sort_values(by="timestamp")

test_data["timestamp"] = pd.to_datetime(test_data["timestamp"])
test_data = test_data.sort_values(by="timestamp")
```

```python
train_data.head()
```

```python
features = ["P1_FCV01D", "P1_FCV01Z", "P1_FCV03D", "P1_FT01"]
target = "x1003_24_SUM_OUT"

# Normalize the data
scaler = MinMaxScaler()
train_data[features] = scaler.fit_transform(train_data[features])
test_data[features]  = scaler.transform(test_data[features])

# Convert to numpy arrays
TrainX = train_data[features].values
TrainY = train_data[target].values

TestX = test_data[features].values
TestY = test_data[target].values
```

```python
print(f"{'Dataset':<10}{'Shape'}")
print(f"{'-'*20}")
print(f"{'TrainX':<10}{TrainX.shape}")
print(f"{'TrainY':<10}{TrainY.shape}")
print(f"{'TestX':<10}{TestX.shape}")
print(f"{'TestY':<10}{TestY.shape}")
```

```python
def create_sequences(data, targets, seq_length):
    sequences = []
    labels = []
    for i in range(len(data) - seq_length):
        seq = data[i:i + seq_length]
        label = targets[i + seq_length]
        sequences.append(seq)
        labels.append(label)
    return torch.tensor(sequences, dtype=torch.float32), torch.tensor(labels, dtype=torch.float32)

SEQ_LENGTH = 3
X_train, y_train = create_sequences(TrainX, TrainY, SEQ_LENGTH)
X_test,  y_test  = create_sequences(TestX,  TestY,  SEQ_LENGTH)
```

```python
print(f"{'Dataset':<10}{'Shape'}")
print(f"{'-'*20}")
print(f"{'X_train':<10}{X_train.shape}")
print(f"{'y_train':<10}{y_train.shape}")
print(f"{'X_test':<10}{X_test.shape}")
print(f"{'y_test':<10}{y_test.shape}")
```

```python
train_loader = DataLoader(list(zip(X_train, y_train)), batch_size=2, shuffle=True)
test_loader = DataLoader(list(zip(X_test, y_test)), batch_size=2, shuffle=False)
```

### Model Definition

```python
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        out, _ = self.lstm(x)
        out = self.fc(out[:, -1, :])  # Take the last output
        return out.squeeze()

# Model hyperparameters
INPUT_SIZE = len(features)
HIDDEN_SIZE = 64
NUM_LAYERS = 2
LEARNING_RATE = 0.001
EPOCHS = 100

model = LSTMModel(INPUT_SIZE, HIDDEN_SIZE, NUM_LAYERS)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)
```

```python
losses = []
for epoch in range(EPOCHS):
    model.train()
    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()
        y_pred = model(X_batch)
        loss = criterion(y_pred, y_batch)
        loss.backward()
        optimizer.step()

    losses.append(loss.item())
    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{EPOCHS}], Loss: {loss.item():.4f}")

# Plot the training loss
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.show()
```
