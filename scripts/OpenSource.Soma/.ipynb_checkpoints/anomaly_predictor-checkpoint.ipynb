# Load Database

```python
# Define the path to your SQLite database file
database_path = "hai-dataset.db"

connection = sqlite3.connect(database_path)

query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql_query(query, connection)

print("Tables in the database:")
print(tables)
```

### Load tabels and import them as dataframes

```python
train_data = pd.read_sql_query(f"SELECT * FROM train;", connection)
test_data = pd.read_sql_query(f"SELECT * FROM test;", connection)
connection.close()
```

# Normalization

```python
X_train = train_data[numeric_columns]
X_test = test_data[numeric_columns]
y_test = test_data.attack_label
```

```python
X_train.shape, X_test.shape, y_test.shape
```

```python
X_train = (X_train - X_train.min()) / (X_train.max() - X_train.min())
X_test = (X_test - X_test.min()) / (X_test.max() - X_test.min())
```

```python
X_train.describe()
```

```python
X_test.describe()
```

# Clustering Model Development

```python
chosen_columns = [0, 2, 4]

fig, axes = plt.subplots(1, 2, figsize=(15, 7), subplot_kw={'projection': '3d'})

axes[0].scatter(
    X_train.iloc[:, chosen_columns[0]].values,
    X_train.iloc[:, chosen_columns[1]].values,
    X_train.iloc[:, chosen_columns[2]].values,
    marker='o',
    cmap='viridis',
    label="Train"
)
axes[0].set_title("Training Data")
axes[0].legend()

scatter_test = axes[1].scatter(
    X_test.iloc[:, chosen_columns[0]].values,
    X_test.iloc[:, chosen_columns[1]].values,
    X_test.iloc[:, chosen_columns[2]].values,
    c=y_test,
    marker='o',
    cmap='viridis',
    label="Test"
)
axes[1].set_title("Testing Data")
axes[1].legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
model = KMeans(n_clusters=2, random_state=42)
model.fit(X_train.iloc[:, chosen_columns])
```

```python
train_clusters = model.predict(X_train.iloc[:, chosen_columns])
train_preds = [1 if label == 1 else 0 for label in train_clusters]

test_clusters = model.predict(X_test.iloc[:, chosen_columns])
test_preds = [1 if label == 1 else 0 for label in test_clusters]

print(confusion_matrix(y_test, test_preds))
print(classification_report(y_test, test_preds))
```

```python
colors = np.array(['red', 'green'])
i, j = 0, 2
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].scatter(X_train.iloc[:, i].values, X_train.iloc[:, j].values, c=train_preds, cmap='viridis', label="Train, Pred")
axes[1].scatter(X_test.iloc[:, i].values,  X_test.iloc[:, j].values,  c=test_preds,  cmap='viridis', label="Test, Pred")
axes[2].scatter(X_test.iloc[:, i].values,  X_test.iloc[:, j].values,  c=y_test,      cmap='viridis', label="Test, Ground")
plt.legend()
plt.show()
```

```python
train_data["predictions"] = train_preds
test_data["predictions"] = test_preds
test_data.head()
```

# Save to the database

```python
database_path = "data.db"
connection = sqlite3.connect(database_path)

train_data.to_sql('train', connection, if_exists='replace', index=False)
test_data.to_sql('test', connection, if_exists='replace', index=False)

connection.commit()
connection.close()

print("Data saved successfully into the database.")
```
